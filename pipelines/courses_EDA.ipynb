```python
import pandas as pd
import json
```

```python
with open('data/all_courses_data.json', "r") as jsonFile:
    all_courses_data = json.load(jsonFile)['courses']

courses_df = pd.DataFrame(all_courses_data)
courses_df.head()
```

```python
courses_df.columns
```

```python
# number of universities
len(courses_df.nomeStruttura.value_counts())
```

```python
def capitalize_name(name):
    # Lowercase words that should not be capitalized
    lowercase_words = {'della', 'di', 'e', 'con', 'per', 'dell', 'degli', 'del', 'a', 'da', 'in', 'su', 'per', 'tra', 'fra'}
    words = name.split()
    capitalized_words = [words[0].capitalize()] + [word.capitalize() if word.lower() not in lowercase_words else word.lower() for word in words[1:]]
    return ' '.join(capitalized_words)

courses_df['nomeStruttura'] = courses_df.nomeStruttura.apply(capitalize_name)
courses_df.nomeStruttura.value_counts()
```
